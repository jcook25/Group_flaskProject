```python
data[['latitude','longitude']] = data._geoloc.apply( 
   lambda x: pd.Series(str(x).split(", "))) 
   
data.head(3)
```

```python
df = data.loc[:,['trail_id', 
                 'name', 
                 'city_name', 
                 'state_name',
                 'country_name',
                 'latitude',
                 'longitude',
                 'popularity',
                 'length',
                 'elevation_gain',
                 'difficulty_rating',
                 'route_type', 
                 'avg_rating',
                 'num_reviews',
                 'features',
                 'activities']]
df.head(3)
```
